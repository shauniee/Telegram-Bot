```python
import random
from pathlib import Path

cwd = Path.cwd()

pic_in_dir = cwd/'Pictures'
pic_list = []
for pic in pic_in_dir.iterdir():
    pic_str = str(pic)
    pic_list.append(pic_str)
random_pic = random.choice(pic_list)
print(random_pic, 'hi')
```
